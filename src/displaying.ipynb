```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import fneuron as fn
import perceptron as pct
import mlp as mlp
import view as vw
```

```python
print('PERCEPTRON (one layer) with heaviside')
w1 = np.array([-10, 1, 0])
w2 = np.array([45, -1, -1])
w3 = np.array([-10, 0, 1])
p = pct.Perceptron(2, 3, [fn.Fneuron(2, w1), fn.Fneuron(2, w2), fn.Fneuron(2, w3)])
data = vw.DataSet2D(p.heaviside, 0, 50, 1000)
view = vw.View(data.data)
view.__show__()
```

```python
print('PERCEPTRON (one layer) with soft-max')
w1 = np.array([-10, 1, 0])
w2 = np.array([45, -1, -1])
w3 = np.array([-10, 0, 1])
p = pct.Perceptron(2, 3, [fn.Fneuron(2, w1), fn.Fneuron(2, w2), fn.Fneuron(2, w3)])
print(p.softmax(np.array([20, 20])))
```

```python
print('PERCEPTRON (multi layers) simulation')
w1 = np.array([-10, 1, 0])
w2 = np.array([45, -1, -1])
w3 = np.array([-10, 0, 1])
p = pct.Perceptron(2, 3, [fn.Fneuron(2, w1), fn.Fneuron(2, w2), fn.Fneuron(2, w3)])
data = vw.DataSet2D(p.heaviside, 0, 50, 1000)
view = vw.View(data.data)
view.test(7)
```

```python
print('XOR logic with Mlp class')
w1 = np.array([-1, 2, 2]) # OR logic
w2 = np.array([3, -2, -2]) # NAND logic
w3 = np.array([-3, 2, 2]) # AND logic
network = mlp.Mlp(2, 1, [2, 1], [[fn.Fneuron(2, w1), fn.Fneuron(2, w2)],[fn.Fneuron(2, w3)]])
print('FAUX et FAUX | output -> ', network.compute(np.array([0, 0])))
print('VRAI et FAUX | output -> ', network.compute(np.array([1, 0])))
print('FAUX et VRAI | output -> ', network.compute(np.array([0, 1])))
print('VRAI et VRAI | output -> ', network.compute(np.array([1, 1])))
```
