```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import fneuron as fn
import perceptron as pct
import mlp
import view as vw
```

```python
print('PERCEPTRON (one layer)')
p = pct.Perceptron(2, 3, np.array([[-10, 1, 0],[45, -1, -1],[-10, 0, 1]]))
data = vw.DataSet2D(p.heaviside, 0, 50, 1000)
view = vw.View(data.data)
view.__show__()
```

```python
print('PERCEPTRON (multi layers) simulation')
p = pct.Perceptron(2, 3, np.array([[-10, 1, 0],[45, -1, -1],[-10, 0, 1]]))
data = vw.DataSet2D(p.heaviside, 0, 50, 1000)
view = vw.View(data.data)
view.test(7)
```

```python
print('XOR logic with Mlp class')
w1 = np.array([-1, 2, 2]) # OR logic
w2 = np.array([3, -2, -2]) # NAND logic
w3 = np.array([-3, 2, 2]) # AND logic
network = mlp.Mlp(2, 1, [2, 1], [[fn.Fneuron(2, w1), fn.Fneuron(2, w2)],[fn.Fneuron(2, w3)]])
print('FAUX et FAUX | output -> ', network.aggregation(np.array([0, 0]))[-1])
print('VRAI et FAUX | output -> ', network.aggregation(np.array([1, 0]))[-1])
print('FAUX et VRAI | output -> ', network.aggregation(np.array([0, 1]))[-1])
print('VRAI et VRAI | output -> ', network.aggregation(np.array([1, 1]))[-1])
```
