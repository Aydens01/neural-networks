```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import fneuron as fn
import perceptron as pct
import view as vw
```

```python
print('PERCEPTRON (one layer)')
p = pct.Perceptron(2, 3, np.array([[-10, 1, 0],[45, -1, -1],[-10, 0, 1]]))
data = vw.DataSet2D(p.heaviside, 0, 50, 1000)
view = vw.View(data.data)
view.__show__()
```

```python
print('PERCEPTRON (multi layers)')
p = pct.Perceptron(2, 3, np.array([[-10, 1, 0],[45, -1, -1],[-10, 0, 1]]))
data = vw.DataSet2D(p.heaviside, 0, 50, 1000)
view = vw.View(data.data)
view.test(7)
```
